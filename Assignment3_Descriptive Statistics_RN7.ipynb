# IRIS Dataset to display some basic statistical details

```python
# read the iris.csv dataset
df1=pd.read_csv("Iris.csv")
```

```python
#displayed first five columns of iris dataset using syntax head()
df1.head()
```

```python
#displayed last five columns of iris dataset using syntax tail()
df1.tail()
```

```python
# displayed mean of column 'species' in iris dataset
df1.groupby('Species').mean()
```

```python
# displayed median of column 'species' in iris dataset
df1.groupby('Species').median()
```

```python
# displayed mode of column 'species' in iris dataset
df1['Species'].mode()
```

```python
# displayed standard deviation of column 'species' in iris dataset
df1.groupby('Species').std()
```

```python
# displayed 10th percentile of column 'species' in iris dataset

df1.groupby('Species').quantile(0.1)
```

```python
# displayed median percentile of column 'species' in iris dataset
df1.groupby('Species').quantile(0.5)
```

```python
# displayed 90th percentile of column 'species' in iris dataset
df1.groupby('Species').quantile(0.9)
```
